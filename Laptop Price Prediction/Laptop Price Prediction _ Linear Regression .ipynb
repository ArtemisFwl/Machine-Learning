We have to make a model predict the price of laptop

Steps involved in the program/Code:

1) Import the libraries

2) Load the dataset

3) Check the dataset

4) Check for missing values

5) Change the categorical values to numerical format using encoder

6) Scaling

8) Train the model

9) test the model

```python
#1) Import the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Rest of the libraries we will import as we need them
```

```python
#Load the dataset

df=pd.read_csv("/content/drive/MyDrive/AIML Programs /Laptop Price Prediction /laptop_price.csv")
```

```python
#Checking the data

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#Checking for the null values

df.isnull().sum()
```

```python
# We got lucky in this database... There are no null values
```

```python
#Let us Now change the categorical values into numericals
#We will do this using get dummies
```

```python
#List of categorical variables

categorical_variables = df.select_dtypes(include=['object']).columns
```

```python
df_encoded=pd.get_dummies(df, columns= categorical_variables, drop_first=True)
```

```python
df_encoded.shape
```

```python
df_encoded.head()
```

```python
# Now let us seperate the target variable

X= df_encoded.drop(columns=['Price (Euro)'])
y=df_encoded['Price (Euro)']
```

```python
# Next step is scaling

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

X_scaled= sc.fit_transform(X)

#Scaling is done
```

```python
#Now we need to split our database into 2 parts : Train and test
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X_scaled, y,
                                                   test_size=0.2,
                                                   random_state= 32)
```

```python
#Apply linear Regression

from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X_train, y_train)
```

```python
#The model has been trained
```

```python
#Now we will make the predictions

y_predict = lr.predict (X_test)
```

```python
# Now we will evaluate the model

from sklearn.metrics import mean_squared_error, r2_score

mse=mean_squared_error(y_test, y_predict)
r2= r2_score(y_test, y_predict)
```

```python
print('mse: \n',mse)
print('r2: \n',r2)
```
