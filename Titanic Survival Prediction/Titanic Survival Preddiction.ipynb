## Checking and handling for missing Values

```python
df.isna().sum()
```

We observe the following things:

1) Age column has missing values

2) Cabin Columns has missing values

3) Embarked column has missing values
____________________________________________

---


In the age column, we have to fill the missing values by mean, median or mode.

In the cabin column, more than 80% data is missing. Thus we will drop that column altogether

In the embarked column, only 2 rows have missing values. Since it is a small qty, we will drop the rows which contain the missing values.

```python
#Drop the cabin columns

df.drop(["Cabin"], axis=1, inplace=True)
```

```python
# Replace missing values of Age with mean value

df["Age"].fillna(df["Age"].mean(), inplace=True)
```

```python
# Drop the 2 rows in Embarked that contains the missing values

df.dropna(subset=["Embarked"], inplace=True)
```

```python
#Now lets check the missing values

df.isna().sum().sum()
```

```python
df.shape
```

2 rows and one column has been dropped

## Dropping the non usefull features

Looking at the dataset we can conclude that features such as Passenger id, Name, ticket will not contribute to the model training.


Thus we can drop those columns.

```python
df.drop(["Name","PassengerId", "Ticket"], axis=1, inplace=True)
```

```python
df.head()
```

## Handling the Categorical Values

We will use one hot encoding to convert categorical values in numerical format.

Sex and Embarked are the 2 columns that are object type. These are the ones where one hot encoding is required.

```python
df["Embarked"].value_counts()
```

```python
df=pd.get_dummies(df, columns=["Sex", "Embarked"], drop_first=True).astype(int)
```

## Feature Engineering

To demonstrate the feature engineering, we will make a new feature: Family Size


```python
df["Family_Size"]=df["SibSp"]+df["Parch"]

#Sibsp: Siblings and spouse
#Parch: Parents
```

## Splitting the dataset

```python
X=df.drop(["Survived"], axis=1)
y=df["Survived"]
```

Here we have seperated the target variable from the Features

## Split dataset into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=43)
```

```python
X_train.shape
```

## Model Selection

We will use different models and check the accuracy with each of them

1) Logistic Regression

2) Decision Tree

3) Random Forest

4) SVM


## 1) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_test_prediction= lr.predict(X_test)
```

```python
# Now we need to evaluate the model
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print("Accuracy Score:\n", accuracy_score(y_test, y_test_prediction))
print("\n\n Confusion Matrix:\n\n", confusion_matrix(y_test, y_test_prediction))
print("\n\nClassification Report:\n", classification_report(y_test, y_test_prediction))

```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
y_test_prediction_dt = dt.predict(X_test)
```

```python
print("Accuracy Score:\n", accuracy_score(y_test, y_test_prediction_dt))
print("\n\n Confusion Matrix:\n\n", confusion_matrix(y_test, y_test_prediction_dt))
print("\n\nClassification Report:\n", classification_report(y_test, y_test_prediction_dt))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)

y_test_prediction_rf=rf.predict(X_test)
```

```python
print("Accuracy Score:\n", accuracy_score(y_test, y_test_prediction_rf))
print("\n\n Confusion Matrix:\n\n", confusion_matrix(y_test, y_test_prediction_rf))
print("\n\nClassification Report:\n", classification_report(y_test, y_test_prediction_rf))
```

## SVM: Support Vector Machine

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train, y_train)

y_test_prediction_svc=svc.predict(X_test)
```

```python
print("Accuracy Score:\n", accuracy_score(y_test, y_test_prediction_svc))
print("\n\n Confusion Matrix:\n\n", confusion_matrix(y_test, y_test_prediction_svc))
print("\n\nClassification Report:\n", classification_report(y_test, y_test_prediction_svc))
```

```python
#Thus we have the accuracy score from 4 different models.

#We observe that Logistic regression gave the highest accuracy
```

## Demonstration of Hyperparameter Tuning

We will do the Hyperparameter Tuning of the Model Random Forest

```python
from sklearn.model_selection import RandomizedSearchCV

grid_parameters={
    "n_estimators":[100,200],
    "max_depth":[1,2,4,5,7,9,10],
    "min_samples_split":[2,4,6,8,10],
    "min_samples_leaf":[1,3,5,7,9]
}

rscv= RandomizedSearchCV(estimator=rf,
                         param_distributions=grid_parameters,
                         cv=5)
rscv.fit(X_train, y_train)
```

```python
best_model=rscv.best_estimator_
```

```python
rscv.best_params_
```

```python
rscv.best_score_
```

We observe that there is slight imporovement in accuracy of Random Forest after hyperparameter.

THough this increase is not so significant.

## Checking for overfititng: Logistic Regression Model

```python
#Check the training data accuracy

train_accuracy= accuracy_score(y_train, lr.predict(X_train))
test_accuracy=accuracy_score(y_test, lr.predict(X_test))

print("train_accuracy:", train_accuracy)
print('test_accuracy',test_accuracy)
```

```python
#Here we see that our model has underfitted.
```

```python
#lets check overfitting in randomForest

train_accuracy= accuracy_score(y_train, rf.predict(X_train))
test_accuracy=accuracy_score(y_test, rf.predict(X_test))

print("train_accuracy:", train_accuracy)
print('test_accuracy',test_accuracy)
```

Here there is a huge difference in Accuracy of test and train dataset. Hence there is a overfitting here.

Model bheaves well with training data but not able to generalise for testing data.
