```python
df=df.astype({"HomePlanet_Europa": int,
              "HomePlanet_Mars":int,
              "Destination_PSO J318.5-22":int,
              "Destination_TRAPPIST-1e":int})
```

```python
df_train=df[df["Transported"].isnull()==False]
df_test=df[df["Transported"].isnull()==True]
```

```python
X=df_train.drop("Transported", axis=1)
y=df_train["Transported"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X,y,
                                                   test_size=0.3,
                                                   random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train, y_train)
```

```python
#Evaluation
#- Accuracy
#- Confusion Matrix
#- Classification Report
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
def model_eval(model):
  accuracy_test= accuracy_score(y_test, model.predict(X_test))
  print("\033[1mTest Accuracy: \033[0m",accuracy_test )
  accuracy_train= accuracy_score(y_train, model.predict(X_train))
  print("\n\033[1mTrain Accuracy\033[0m: ",accuracy_train )

  print("\033[1mDifference of train and test accuracy\033[0m", accuracy_train-accuracy_test)
  percentage_difference= (accuracy_train-accuracy_test)/accuracy_train*100
  print("\033[1m\nPercent Difference:\033[0m", percentage_difference)

  if percentage_difference>5:
    print("\n\n\033[1mGreater than 5% difference in accuracy,\n Model Seems overfitted\033[0m")
  else:
    print("\n\n\033[1mLess than 5% difference in accuracy\033[0m")
```

```python
model_eval(lr)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
model_eval(rf)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
model_eval(knn)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train, y_train)
```

```python
model_eval(nb)
```

Here We have underfitting as the model accuracy is not so good.

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(X_train, y_train)
```

```python
model_eval(sv)
```
